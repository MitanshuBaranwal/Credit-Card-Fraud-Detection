Observation : We can see there are 3 distinct values in Item_Fat_Content columns Low Fat, LF, low fat which might mean same parameter. Smae goes for Regular and reg

```python
def Item_Fat_Content_func(x):
    if x == "low fat" or x == "Low Fat" or x =="LF":
        return 0
    elif x == "Regular" or "reg":
        return 1
    
```

```python
df.Item_Fat_Content = df.Item_Fat_Content.apply(Item_Fat_Content_func)
df
```
